## Definition:


let be $X$ a continous $D$ dimensional variable, and $(x_{1}, t_{1}), (x_{2}, t_{2}), ...(x_{N}, t_{N})$ $N$ observations, $regression$ is a $supervised$ technique(due we have labels or targets $t_{1}, ...t_{N})$ that allows us the predict the value $t$ for a new giveen input $x$.(e,g $polynomial-regression$)

## Linear Basis Functions:


$linear-regression$ is the simplest version of $regression$ as we predict the output $t$($y$ is the predict value of $t$) as the $linear$ combinaison of the input $x$.

$y(x, w)=w_{0}+w_{1}\phi_{1}(x)+...+w_{D}\phi_{D}(x)$

$w_{0}$ is called $bias$.

$w_{0}, w_{1}, ...,w_{D}$ are called the parameters.

$\phi=(\phi_{1}, \phi_{2}, ...,\phi_{D})$ is called the $basis$ function.

for $\phi_{j}(x)=x^{j}$ we obtain a $polynomial-regression$ model.


## Ploting The Basis Functions:

we consider here the following basis function:

1-polynomial basis function:

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~$ $\phi_{j}(x)=x^{j}$

2-gussian basis function:

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~$ $\phi_{j}(x)=\exp(-(x_{j}-\mu_{j})^2/2s^{2})$

3-the sigmoidable basis function:

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~$ $\phi_{j}(x)=\sigma((x_{j}-\mu_{j})/s)$

where $\sigma(.)$ is a sigmoid function giveen by:

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~$ $\sigma(a)=1/(1+\exp(-a))$

```python
import numpy as np
import matplotlib.pyplot as plt

def polynomial_bas(x, j):
   return x**j

def gaussian_bas(x, j, mu_j=1., s=.25):
  return np.exp(-(x-mu_j)**2/(2*s**2))

def sigmoid(a):
  return 1/(1+np.exp(-a))

def sigmoidable_bas(x, j, mu_j=1., s=0.25):
  return sigmoid(np.exp((x-mu_j)/s))

x=np.linspace(-1, 1, 1000)
for j in range(1, 10):
    y=polynomial_bas(x, j)
    plt.plot(x, y)
plt.show()

for j in range(1, 10):
    for mu_j in np.linspace(-1, 1, 10):
      y= gaussian_bas(x, j, mu_j)
      plt.plot(x, y, )
plt.show()

for j in range(1, 10):
  for mu_j in np.linspace(-1, 1, 10):
      y=sigmoidable_bas(x, j, mu_j)
      plt.plot(x, y)
plt.show()




```
