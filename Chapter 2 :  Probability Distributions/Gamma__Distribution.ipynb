## Gamma distribution:

the $ gamma $ distribution is a $conjugute$ of  $posterior$  $gaussian$ distribution with $unkown$ variance($\sigma^2$) and $mean$ $\mu$ and it's giveen by the following expression:

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~$ $ Gam(\lambda /a,b)=\dfrac{1}{\gamma(a)}b^a\exp(-\lambda b)$

where $\lambda=\dfrac{1}{\sigma^2}$

```python
import numpy as np
# import base function gamma
from numpy.random import gamma as gamma_fun
import matplotlib.pyplot as plt

class gamma:

    def __init__(self, a:float, b:float):
        """ a and b are the parameters that controls the shape of distribution """
        self.a=a
        self.b=b

    def pdf(self, lamda):
        return self.b**self.a*np.exp(-lamda*self.b)*lamda**(self.a-1)/gamma_fun(self.a)


a_list, b_list=[0.1, 1, 4], [0.1, 1, 6]
lamda=np.linspace(0,1,100)
for a , b in zip(a_list, b_list):
   gam=gamma(a=a, b=b)
   y=gam.pdf(lamda)
   plt.plot(lamda, y, color='red', label='a='+str(a)+'  '+'b='+str(b))
   plt.legend()
   plt.show()
    


    

      

     

```
