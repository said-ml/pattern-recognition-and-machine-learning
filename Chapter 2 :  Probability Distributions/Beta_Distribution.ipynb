## Beta's Distribution:


the $beta$ distribution is the $conjugate$ distribution of the $binomial$ distribution and it is giveen by the following expression:

$~~~~~~~~~~~~~~~~$ $Beta(\mu/a,b)=\dfrac{\gamma(a+b)}{\gamma(a)\gamma(b)}\mu^{a-1}(1-\mu)^{b-1}$

$a$ and $b$ are the $hyperparameters$ that control the $Beta$ distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

liste_a, liste_b=[0.1, 1, 2, 8], [0.1, 1, 3, 4]
mu = np.linspace(0, 1, 100)
for a, b in zip(liste_a, liste_b):
    beta= mu**(a-1) * (1-mu)**(b-1) / (np.math.gamma(a) * np.math.gamma(b) / np.math.gamma(a+b))
    plt.plot(mu,beta, label='a='+str(a)+"  "+'b='+str(b), color='red')
    plt.xlabel('$ \mu $')
    plt.legend()
    plt.show()
```

## Prior, Likelihood, Posterior:

let's be $\mathbb{D}$ the data, and $\mu$ an parameter(exactly hyperparmeter), if we take advantage of $Bayes$ formula, we obtain:

$~~~~~~~~~~$ $p(\mu/\mathbb{D})=\dfrac{p(\mu)p(\mathbb{D}/\mu)}{p(\mathbb{D})}$ $\star$

$p(\mu)$ is called a $prior$ distribution (no $data$~$point$ is giveen).

$p(\mathbb{D}/\mu)$ is called $likelihood$ function, it's the distribution of data given the $hyperparameter$(not necessary normalized).

${p(\mathbb{D})}$ is called $  evidence$ , is constant over $\mu$(independant of $\mu$)

$p(\mu/\mathbb{D})$ is called $posterior$ distribution , it's the $inference$ of the $hyperparameter$ $\mu$ giveen $data$  $\mathbb{D}$, and in general is the target of the $bayesian$ $inference$ problem.

in our case we have our $data$ is $N=m=1$ and the $likelihood$ function is giveen by:

$~~~~~~~~~~~~~~~$$likelihood=Bin(1/ 1,\mu)=\mu$

and the $prior$ is giveen by:

$~~~~~~~~~~~~~~~$$prior=Beta(\mu/a=2, b=2)$

such as :

$~~~~~~~~~~~~~~~$$posterior\approx prior \times likelihood$

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~$ $\approx \mu^{2-1}(1-\mu)^{2-1}\mu(a=b=2)$

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~$ $\approx \mu^{3-1}(1-\mu)^{2-1}$

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~$ $\approx Beta(\mu /3,2) (a=3 ,  b=2)$



```python
import numpy as np

mu=np.linspace(0,1,1000)
def beta(a, b, x):
    return mu**(a-1) * (1-mu)**(b-1) / (np.math.gamma(a) * np.math.gamma(b) / np.math.gamma(a+b))

prior=beta(2,2,mu)
plt.plot(mu,prior, label='prior', color='red')
plt.xlabel('$ \mu $')
plt.legend()
plt.show()

plt.plot(mu,mu, label='likelihood function', color='blue')
plt.xlabel('$ \mu $')
plt.legend()
plt.show()

posterior=beta(3,2,mu)
plt.plot(mu,posterior, label='posterior', color='red')
plt.xlabel('$ \mu $')
plt.legend()
plt.show()

```
