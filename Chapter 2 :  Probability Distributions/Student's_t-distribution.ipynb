## Student's t-distribution:

the $ Student$'s t-distribution is giveen by the following expression:

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~$$St(x/\mu,\lambda, \nu)=\dfrac{\gamma(\nu/2+1/2)}{\gamma(\nu/2)}(\lambda/\pi\nu)^{1/2}[1+\lambda(x-\mu)^2/\nu)]^{-\nu/2-1/2}$

```python
import numpy as np
from scipy.special import gamma as gamma_fun
import matplotlib.pyplot as plt 

class Student:

    def __init__(self, mu=0, lamda=1, nu=.1):
        self.mu=mu
        self.lamda=lamda
        self.nu=nu

    def pdf(self, x):
        return (gamma_fun(self.nu/2+1/2)/gamma_fun(self.nu/2))*(self.lamda/(np.pi*self.nu))**0.5*(1+self.lamda*(x-self.mu)**2*1/self.nu)**(-self.nu/2-0.5)


x=np.linspace(-5, 5, 100)
for nu in [.1, 1., 200]:
    # we set n=200 as infinity, np.inf doesn't work
    std=Student(nu=nu)
    y=std.pdf(x)
    plt.plot(x, y, label=r'$\nu=$'+' '+str(nu))
    plt.legend()
plt.show()
    
```

as we can see when $\nu$ is getting large, the $ Student$'s t-distribution is tend towardsthe $\mathbb{N(./\mu, \lambda^{-1})}$ distribution where $\lambda=\dfrac{1}{\sigma^2}$ is the $precision$ of $\mathbb{N(./\mu, \lambda^{-1})}$.
