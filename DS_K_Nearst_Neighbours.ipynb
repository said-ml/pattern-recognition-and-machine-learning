```python
from ipynb.fs.full.DS_plot_state_borders import plot_state_borders
Vector=List[float]

def raw_majority_vote(labels:List[str])->str:
    votes=Counter(labels)
    winner, winner_count=votes.most_common(1)[0]
    return winner

liste=['A','B','C','C']
assert raw_majority_vote(liste)=='C'

def majority_vote(labels:List[str])->str:
    votes=Counter(labels)
    winner, winner_count=votes.most_common(1)[0]
    num_winner=len([count for count in votes.values()
                   if count==winner_count])
    if num_winner==1:
        return winner
    else:
        return majority_vote(labels[:-1])
 
liste=['A','B','C','C','B']
assert majority_vote(liste)=='C'

from typing import NamedTuple, List
from ipynb.fs.full.DS_Linear_Algebra import distance

class LabelPoint(NamedTuple):
    point:Vector
    label:str
# cities=[(latitude, longtitude), programming_language]

def knn_classify(k:int, labeled_points:List[LabelPoint], new_point:Vector)->str:
    '''each scale point is a pair of scale and point'''
    # ordered the labels from the nearest to the farthest
    by_distance=sorted(labeled_points, key=lambda lp:distance(lp.point, new_point))
    # find the labels for the k closet
    k_nearst_labels=[lp.label for lp in by_distance[:k]]
    # let them vote
    return majority_vote(k_nearst_labels)

import random

def random_point(dim:int)->Vector:
    return [random.random() for _ in range(dim)]

def random_distances(dim:int, num_pairs:int)->List[float]:
    return [distance(random_point(dim), random_point(dim)) for _ in range(num_pairs)]

def main():
    from typing import Dict
    import csv
    from collections import defaultdict
    def parse_iris_row(row:List[str])->LabelPoint:
        measurements=[value for value in row[-1]]
        label=row[-1].split('-')[-1]
        return LabelPoint(measurements, label)
    with open('/home/said/Downloads/data-science-from-scratch_comma_delimited_stock_prices_files/iris.csv') as f:
        reader=csv.reader(f)
        iris_dataa=[row for row in reader]
    iris_data=[]
    for row in iris_dataa[1:]:
        a, b, c, d, e=row
        a, b, c, d,e=float(a), float(b), float(c), float(d), e
        row=a, b, c, d, e
        iris_data.append(row)
    iris_data=[parse_iris_row(row) for row in reader]
    
        
   

     # we'll also group just the points by species / label we can plot them 
    point_by_species:Dict[str, List[Vector]]=defaultdict(list)
    for iris in iris_data:
         point_by_species[iris.label].append(iris.point)
            
    import matplotlib.pyplot as plt
    metrics=['sepal_length', 'sepal_with', 'petal_length', 'petal_length']
    pairs=[(i,j) for i in range(4) for j in range(4) if i<j]
    marks=['+','.','x'] # we have three classes so we have three markers
    ax, fig=plt.subplots(2,3)
    for row in range(2):
        for col in range(3):
            i,j=pairs[3*row+col]
            ax[row][col].set_title(f"{metrics[i]} vs {metrics[j]}", fontsize=8)
            ax[row][col].set_xticks([])
            ax[row][col].set_yticks([])
            for mark ,(species, points) in zip(marks, point_by_species):
                  xs=[point[i] for pointt in points]
                  ys=[point[j] for j in points]
                  plt.scatter(xs, ys, marker=mark, label=species)
    ax[-1][-1].legend(loc='lower right', prop={'size', 6})
    plt.show()
if __name__=='__main__':main()
        


    












```
